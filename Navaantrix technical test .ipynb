### 2. Read

```python
def read_entry(updated_amazon, product_id):
    return updated_amazon[updated_amazon["product_id"] == product_id]

# Example
product_details = read_entry(updated_amazon, "B09example1")

# Parameters:
# updated_amazon: A pandas DataFrame containing the sales data.
# product_id: The ID of the product that you want to retrieve from the DataFrame.

# Return:
#The function returns a DataFrame containing the rows where the product_id matches the specified product_id.
```

### 3. Update

```python
def update_entry(updated_amazon, product_id, column, new_value):
    updated_amazon.loc[updated_amazon["product_id"] == product_id, column] = new_value
    return updated_amazon

# Example
updated_amazon = update_entry(updated_amazon, "B09example1", "discounted_price", "₹000")

# Parameters:
# updated_amazon: A pandas DataFrame containing the sales data.
# product_id: The ID of the product that you want to update.
# column: The column name that you want to update.
# new_value: The new value to set for the specified column.

# Return:
# The function returns the updated DataFrame with the specified column value updated for the matching product_id.
```

### 4. Delet

```python
def delete_entry(updated_amazon, product_id):
    return updated_amazon[updated_amazon["product_id"] != product_id]
    # This line filters the DataFrame, keeping only the rows where the product_id does not match the given product_id

# Example
updated_amazon = delete_entry(updated_amazon, "B09example5")

# Parameters:
# updated_amazon: A pandas DataFrame containing the sales data.
# product_id: The ID of the product that you want to remove from the DataFrame.

# Return:
# The function returns a new DataFrame with the specified product_id removed.
```
